```python
from tkinter import *
import math
from pygame import mixer
import speech_recognition
```

```python
def click(val):
    x = showfield.get()
    ans = ''
    
    try:
        if val == 'C':
            x = showfield.get()
            x = x[0:len(x)-1]
            showfield.delete(0, END)
            showfield.insert(0, x)
            return
        elif val == 'CE':
            showfield.delete(0, END)
        elif val == '√':
            ans = math.sqrt(eval(x))
        elif val == 'π':
            ans = math.pi
        elif val == 'cosθ':
            ans = math.cos(math.radians(eval(x)))
        elif val == 'tanθ':
            ans = math.tan(math.radians(eval(x)))
        elif val == 'sinθ':
            ans = math.sin(math.radians(eval(x)))
        elif val == '2π':
            ans = 2*math.pi
        elif val == 'cosh':
            ans = math.cosh(eval(x))
        elif val == 'tanh':
            ans = math.tanh(eval(x))
        elif val == 'sinh':
            ans = math.sinh(eval(x))
        elif val == chr(8731):
            ans = eval(x)**(1/3)
        elif val == 'x\u02b8':
            showfield.insert(END, '**')
            return
        elif val == 'x\u00B3':
            ans = eval(x)**3
        elif val == 'ln':
            ans = math.log2(eval(x))
        elif val == 'deg':
            ans = math.degrees(eval(x))
        elif val == 'rad' :
            ans = math.radians(eval(x))
        elif val == 'e' :
            ans = math.e
        elif val == 'log10':
            ans = math.log10(eval(x))
        elif val == 'x!':
            ans = math.factorial(x)
        elif val == chr(247):
            showfield.insert(END,'/')
            return
        elif val == '=':
            ans = eval(x)
        else:
            showfield.insert(END,val)
            return


        showfield.delete(0, END)
        showfield.insert(0, ans)
    
    except SyntaxError:
        pass
    

    
```

```python
mixer.init()
def add(x,y):
    return x+y
def sub(x,y):
    return x+y
def mul(x,y):
    return x*y
def div(x,y):
    return x/y
def mod (x,y):
    return x%y
def lcm(x,y):
    l = math.lcm(x,y)
    return l
def hcf(x,y):
    h = math.gcd(x,y)
    return h

op = {'ADD': add, 'ADDITION': add, 'SUM': add, 'PLUS': add,
      'SUBTRACTION': sub, 'DIFFERENCE': sub, 'MINUS': sub, 'SUBTRACT': sub, 
      'PRODUCT': mul, 'MULTIPLY': mul, 'MULTIPLICATION': mul, 'INTO': mul,
      'DIVISION' : div, 'DIVIDED BY' : div, 'DIVIDE' : div, 'DIV' : div, 'BY' : div,
      'LCM': lcm, 'HCF': hcf, 'MOD': mod, 'REMAINDER': mod, 'MODULUS': mod}

def nums(x):
    t = []
    for n in x:
        try:
            t.append(float(n))
        except ValueError:
            pass
    return t
           


def audio():
    mixer.music.load('music1.mp3')
    mixer.music.play()
    sp = speech_recognition.Recognizer()
    with speech_recognition.Microphone()as m:
        try:
            sp.adjust_for_ambient_noise(m, duration = 0.2)
            voice = sp.listen(m)
            msg = sp.recognize_google(voice)
            mixer.music.load('music2.mp3')
            mixer.music.play()
            msgs = msg.split('')
            for txt in msgs:
                if txt.upper() in op.keys():
                    k = nums(msgs)
                    res = op[txt.upper()](k[0],k[1])
                    showfield.delete(0, END)
                    showfield.insert(END, res)
                else:
                    pass
                
                    
                    
        except:
            pass
    
```

```python
root = Tk()
root.title('Scientific Calculator')
root.config(bg = 'black')
root.geometry('680x486+100+100')

logo = PhotoImage(file = 'logo.png')
logo_label = Label(root, image =logo, bg = 'black')
logo_label.grid(row = 0, column = 0)


showfield = Entry(root, font=('arial',20,'bold'), bg ='black', fg= 'white', bd= 10, relief = SUNKEN, width =30)
showfield.grid(row = 0, column = 0, columnspan = 8)

mic = PhotoImage(file = 'microphone.png')
mic_button = Button(root, image = mic, bd =0, bg = 'black', activebackground = 'black', command = audio)
mic_button.grid(row = 0, column = 7)

button_list = ["C", "CE", "√", "+", "π", "cosθ", "tanθ", "sinθ", "1", "2", "3", "-", "2π", "cosh",
               "tanh", "sinh", "4", "5", "6", "*", chr(8731), "x\u02b8", "x\u00B3", "x\u00B2", "7", "8", "9", 
               chr(247), "ln", "deg", "rad", "e", "0", ".", "%", "=", "log10", "(", ")", "x!"]
row_val = 1
col_val = 0
for i in button_list:
    button = Button(root, width = 5, height = 2, bd = 2, relief = SUNKEN, text = i, bg = 'black', fg = 'white', font = ('arial',18,'bold'),activebackground='black', command = lambda button = i: click(button))
    button.grid(row = row_val, column = col_val, pady = 1)
    col_val+=1
    if col_val>7:
        row_val+=1
        col_val = 0


root.mainloop()
```
